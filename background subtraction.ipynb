There is a facility to do background subtraction. The function is 'subkg'. Let us see it in use here.

```python
%matplotlib inline 
#this is useful if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

```python
pc.plot('joined_big_sliced_115_500.fit', 'joined_big_sliced_115_500.png', option=3, blevel=127)
```

Let us do the background subtraction and then plot the resultant file.

```python
pc.subkg('joined_big_sliced_115_500.fit', 'joined_big_sliced_115_500_sub.fit')
```

```python
pc.plot('joined_big_sliced_115_500_sub.fit', 'joined_big_sliced_115_500_sub.png', option=3, blevel=0)
```
