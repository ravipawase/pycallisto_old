# Slicing : Along time axis

Similar to the case of 'slicing along frequency axis' we can also slice the fits file along the time axis thus creating a new fits file.<br>

Let us see the example of slicing along Time axis. Here we plot a file and we observe that we are only  interested in a blobby feature at the end of the file, so our aim is get only that part.<br>
To function for this is <b>sliceslicetimeaxis</b>.<br>


You need to provide<br>
1. the name of original fits file and 
2. the new fits file which will be generated after slicing and 
3. the starttime (string) in "HH:MM:SS" format 
4. endtime (string) in "HH:MM:SS" format 

```python
%matplotlib inline 
#this is useful if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

Here we plot the initial fits file for inspection.

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_1.png', option=3, blevel=127)
```

Here we decide starttime should be "03:24:00" and endtime should be "03:27:00", we input this and let us see what is the result.

```python
pc.slicetimeaxis('IISERP_20151104_031152_59.fit', "03:24:00", "03:27:00", 'IISERP_20151104_031152_59_tsliced.fit')
```

Oh ho! We made a mistake, end time is beyond the range of input file, so it gives error and do not complete the slice operation.It also provides the start and end date-time of the input file so one can choose start and end time for slicing wisely.<br>
Ok so let us slice from "03:24:00" to "03:26:51".

```python
pc.slicetimeaxis('IISERP_20151104_031152_59.fit', "03:24:00", "03:26:51", 'IISERP_20151104_031152_59_tsliced.fit')
```

It seems to have no problem thus creating the sliced fits file. Let us plot this newly generated sliced fits file and see what we have achieved.

```python
pc.plot('IISERP_20151104_031152_59_tsliced.fit', 'IISERP_20151104_031152_59_tsliced.png', option=3, blevel=127, xtick= 1)
```

Good, we are successful in whatever we wanted to achieve.
