# Get-timeseries

As the name suggests the function 'gettimeseries' can be used to create a "timeseries" from the input fits file by collapsing on time axis. You can choose to plot it or retutn data or both.<br>
Let us see this by one example.


```python
%matplotlib inline 
#this is useful if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

Here we plot the initial fits file for inspection.

```python
pc.plot('IISERP_20151104_031152_59_sliced.fit', 'IISERP_20151104_031152_59_1.png', option=3, blevel=127)
```

This image shows the detailed info about the signal. But if we are interested in total signal strength in all frequency band and its variation along time axis then we can use 'gettimeseries' function<br>
Let us run it on this image and see the results.<br>
You need to provide<br>
1. the name of original fits file and 
2. tell to plot or not (True or false)
3. tell to return data or not (True or false)

```python
pc.gettimeseries(infits='IISERP_20151104_031152_59_sliced.fit', plot=True, returndata=False )
```

Ok, this looks fine. It agrees with the original image plot where activity has gone up just before time "03:26:00".<br>
But as we see in the original image plot, all the frequencies above 400 MHz do not show any activity. So we are interested in only plotting the activity from 200 Mhz to 400 Mhz. So we can slice the original fits image along frequency axis and then use 'gettimeseries' on the resultant fits image.

Let us slice it.

```python
pc.slicefreqaxis('IISERP_20151104_031152_59_sliced.fit', 200, 400, 'IISERP_20151104_031152_59_sliced_200_400.fit')
```

Ok, just plot it to confirm that slicing is done properly.

```python
pc.plot('IISERP_20151104_031152_59_sliced_200_400.fit', 'IISERP_20151104_031152_59_sliced_200_400.fit.png', option=3, blevel=127)
```

Ok, good, it's fine. Now let us apply 'gettimeseries' on the sliced fits image and ask to plot it.

```python
pc.gettimeseries(infits='IISERP_20151104_031152_59_sliced_200_400.fit', plot=True, returndata=False )
```

Here you can observe the change in the signal level.

Ok, Till now , all is well. But lets be supergreedy and demand for the 'timeseries' of single frequency. Let us say 203 MHz. So for that purpose we will need to slice a single frequency from original fits image and then pass it to 'gettimeseries'.Let us try it.

```python
pc.slicefreqaxis('IISERP_20151104_031152_59_sliced.fit', 202, 203, 'IISERP_20151104_031152_59_sliced_thinner.fit')
```

Ok, this may happen because the frequency bands are not sampled uniformly

```python
pc.slicefreqaxis('IISERP_20151104_031152_59_sliced.fit', 202, 204, 'IISERP_20151104_031152_59_sliced_thinner.fit')
```

```python
pc.plot('IISERP_20151104_031152_59_sliced_thinner.fit', 'IISERP_20151104_031152_59_sliced_thinner.png', option=3, blevel=127)
```

```python
pc.gettimeseries(infits='IISERP_20151104_031152_59_sliced_thinner.fit', plot=True, returndata=False )
```
