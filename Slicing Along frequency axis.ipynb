# Slicing : Along frequency axis

One may need to slice out a small part of the data to remove some noise on the edge of the file or one may be interested in showing only the interesting signal in the file. It can be done with slicing. We have two functions for slicing, one slices a fits file along frequency-axis  and one slices a file along time-axis thus <b>creating new fits file</b>. 

## Slicing along frequency axis

Let us see the example of slicing along frequency axis. Here we plot a file and we observe that there is a systematic noise a the bottom frequencies, so our aim is to chop off that noisy part.<br>
To function for this is <b>slicefreqaxis</b>.<br>


You need to provide<br>
1. the name of original fits file and 
2. the new fits file which will be generated after slicing and 
3. the startfrequency and 
4. endfrequency for chopping.

```python
%matplotlib inline 
#this is useful if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

Here we plot the initial fits file for inspection.

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_1.png', option=3, blevel=127)
```

Here we decide startfrequency should be 125 and endfrequency should be 900, we input this and let us see what is the result.

```python
pc.slicefreqaxis('IISERP_20151104_031152_59.fit', 125, 900, 'IISERP_20151104_031152_59_sliced.fit')
```

Here end-frequency is out of bound(range) so it gives error and do not complete the slice operation. It also provides the start and end frequency of the input file so one can choose start and end frequency for slicing wisely.<br>
Ok so let us slice from 125 to 870.

```python
pc.slicefreqaxis('IISERP_20151104_031152_59.fit', 125, 870, 'IISERP_20151104_031152_59_sliced.fit')
```

It seems to have no problem thus creating the sliced fits file. Let us plot this newly gnerated sliced fits file and see what we have achieved.

```python
pc.plot('IISERP_20151104_031152_59_sliced.fit', 'IISERP_20151104_031152_59_sliced.png', option=3, blevel=127)
```

Good, we are successful in slicing off the bottom systematic noise part.<br>
At this point one may wish to only show the interesting features below 400 MHz. In this case let us slice the original fits image from 125 MHz to 400 Mhz and then plot it.

```python
pc.slicefreqaxis('IISERP_20151104_031152_59.fit', 125, 450, 'IISERP_20151104_031152_59_sliced1.fit')
```

```python
pc.plot('IISERP_20151104_031152_59_sliced1.fit', 'IISERP_20151104_031152_59_sliced1.png', option=3, blevel=127)
```

In this way one can do slicing in frequency axis according to once needs.
