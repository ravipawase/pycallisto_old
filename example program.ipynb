Here is this eaxmple we take fits image, slice off its noisy part at bottom and then slice in time axis to get the only event part.  Then do the background subtraction and then plot the resultant image.<br>
You can observer the difference between the unsubtracted and subtracted image.

```python
%matplotlib inline 
#this is useful if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_1.png', option=3, blevel=127)
```

```python
pc.plot('IISERP_20151104_032652_59.fit', 'IISERP_20151104_032652_59.png', option=3, blevel=127)
```

```python
pc.jointimeaxis('IISERP_20151104_031152_59.fit', 'IISERP_20151104_032652_59.fit', 'joined_big.fits')
```

```python
pc.plot('joined_big.fits', 'joined_big.png', option=3, blevel=127)
```

```python
pc.slicefreqaxis('joined_big.fits', 115, 500, 'joined_big_sliced_115_500.fit')
```

```python
pc.plot('joined_big_sliced_115_500.fit', 'joined_big_sliced_115_500.png', option=3, blevel=-580)
```

```python
pc.subkg('joined_big_sliced_115_500.fit', 'joined_big_sliced_115_500_sub.fit')
```

```python
pc.plot('joined_big_sliced_115_500_sub.fit', 'joined_big_sliced_115_500_sub.png', option=3, blevel=0)
```

```python
pc.slicetimeaxis('joined_big_sliced_115_500_sub.fit', "03:21:00", "03:36:00", 'final.fit')
```

```python
pc.plot('final.fit', 'final.png', option=3, blevel=-5, fontsize=14)
```
