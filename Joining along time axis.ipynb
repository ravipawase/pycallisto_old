# Joining along time axis

One may find in need to join to files in time axis since the typical individual file coming from [callisto](http://www.e-callisto.org/) is of 15 minutes and an event or feature may extend more than that.<br>
In this case <b>jointimeaxis</b> function is useful. To use this you need to provide following inputs
1. the name/path of first input fits file
2. the name/path  of second input fits file
3. the name/path  of final stiched file to be saved.

These input files are supposed to have same dimentions in frequency and should be continuous in time (though a gap of 1 sec is ignored) and should have the same time-sampling these things are checked and if not followed function won't go ahead with joining operation. If your inputs files are from the same observatory you should not bother about these they should match naturally(unless, if you have changed them)<br>

We see an example below.


```python
%matplotlib inline 
#this is useful if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

Here we plot the input fits files for inspection.

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_1.png', option=3, blevel=127)
```

```python
pc.plot('IISERP_20151104_032652_59.fit', 'IISERP_20151104_032652_59.png', option=3, blevel=127)
```

Let us join these two fits files and save on disk.

```python
pc.jointimeaxis('IISERP_20151104_031152_59.fit', 'IISERP_20151104_032652_59.fit', 'joined.fits')
```

Let us plot the <b>joined.fits</b> file.

```python
pc.plot('joined.fits', 'joined.png', option=3, blevel=127)
```

Now, here in this plot one can see a feature, nice in the middle, which was on the edge of the first input fits image.<br>

In this way, ideally, one can join all the data of whole day.<br>

If you further feel a need to slice only the central part of the joined fits image on time axis putting the feature of interest in right middle, one can do so as shown in the slicing example with proper time inputs.<br>
