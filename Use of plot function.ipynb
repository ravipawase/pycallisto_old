# Use of plot function

Use of plot function is straight-forward, give the function the path of the callisto fits image and the path/name of the png image to be stored.<br>
There are several other keyword options which you can use to enhance the plot as you need.
Important keyword is "option", It can have intiger values 1, 2 or 3 and respectively it plots<br>
1 : raw "row count" verses "column count" of the callisto fits data file<br>
2 : "Time (sec of day)" verses "Frequency (MHz)"<br>
3 : "Universal Time (HH:MM:SS)" verses "Frequency (MHz)"<br>
This is expained below

```python
%matplotlib inline 
#this is usefull if you are coding in jupyter notebook, if not neglect it
```

```python
import pyCallisto as pc   #import pyCallisto Module
```

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_1.png', option=1)
#plot the given fits file with option 1
```

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_2.png', option=2)
#plot the given fits file with option 2
```

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_3.1.png', option=3)
#plot the given fits file with option 3
#this is the needed most of the time and it is default
```

In the above image you can observe, most of the interesting features are above the data value 127, except some systematic noise at bottom frequencies.So to improve the visibility / readability of the image you can set background level of data with keyword <b>blevel</b> as shown in the following example.

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_3.2.png', option=3, blevel=130)
```

This image gives a nice representation of your data, name of the input fits file and starting date-time is shown at the top. But the start and end time and frequnecies are not shown on the respective axes. If you want this, you can force this behaviour by keyword <b>endpts</b> which takes a list of two boolean, each for one axis x and y respectively. By default this behaviour is set to False<br>
Depending your data you may get overlap of endpoint labels and intermediate labels.

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_3.3.png', option=3, blevel=130, endpts=[True, True])
```

You can change the default <b>tick-frequency</b> on the x-axis by keyword <b>xtick</b>. It takes an integer value as input, suppose 3, so it will draw x-ticks at every 3 minutes as shown in the following example.

```python
pc.plot('IISERP_20151104_031152_59.fit', 'IISERP_20151104_031152_59_3.4.png', option=3, blevel=130, xtick= 3)
```
